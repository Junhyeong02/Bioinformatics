```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
```

```python
with open("CaCM334.MYB.total.fa") as f:
    gene_list = [line[1:].strip() for line in f.readlines() if line[0] == ">"]
```

```python
df = pd.read_csv("CaCM334.MYB.ReAnnot.chromosome.ko00906.ripening.txt", sep = "\t", comment = "#")
```

```python
df
```

```python
ex_df = pd.read_csv("../expression_DB/source/CM334.tissue.avg.tsv", sep = "\t", index_col = "GeneID")
```

```python
ex_df = ex_df[["RT", "ST", "LF", "FL", "PL1", "PL2", "PL3", "PLMG", "PLB", "PLB5", "PLB10", "PR1", "PR2", "PR3", "PRMG", "PRB", "PRB5", "PRB10"]]
```

```python
my_df = pd.merge(df, ex_df, how = "inner", left_on = "gene", right_index = True)
```

```python
my_df
```

```python
my_df.drop(["name", "ko", "gene"], axis = 1, inplace = True)
```

```python
my_df 
```

```python
my_df.dropna(inplace = True)
```

```python
my_df = my_df.set_index("symbol")
```

```python
my_df = my_df.loc[(my_df!=0).any(axis=1)]
```

```python
my_df.dropna(axis = 1, inplace = True)
```

```python
my_df
```

```python
my_df.to_csv("sample.txt", sep = "\t", index_label = "ID")
```

```python
my_df = my_df.div(my_df.max(axis=1), axis=0)
```

```python
clu = sns.clustermap(my_df, cmap = 'viridis', col_cluster = False, yticklabels = True, cbar_pos=(-0.1, 0.2, .03, .4), 
                     standard_scale = 0, figsize= (20, 15), metric='correlation', dendrogram_ratio = 0.1)

for i, ax in enumerate(clu.fig.axes):   ## getting all axes of the fig object|
    ax.set_xticklabels(ax.get_xticklabels(), fontdict={'fontsize' : 20})
    
        
for t in clu.fig.axes[2].get_yticklabels():
    t.set_fontsize(20)
    
print(clu.fig.axes)

# plt.savefig('MEBrown_Expression.png')
# plt.show()
# plt.savefig('Tree group.jpg')
# plt.savefig('clustering_carotenoid.png', bbox_inches = "tight")
plt.show()
```
