```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
```

```python
with open("CaCM334.MYB.total.fa") as f:
    gene_list = [line[1:].strip() for line in f.readlines() if line[0] == ">"]
```

```python
df = pd.read_csv("CaCM334.MYB.ReAnnot.chromosome.ko00906.txt", sep = "\t")
```

```python
df
```

```python
ex_df = pd.read_csv("../expression_DB/source/CM334.all.TPM.tsv", sep = "\t", index_col = "GeneID")
```

```python
my_df = pd.merge(df, ex_df, how = "inner", left_on = "gene", right_index = True)
```

```python
my_df
```

```python
my_df.drop(["gene", "ko"], axis = 1, inplace = True)
```

```python
my_df 
```

```python
my_df.dropna(inplace = True)
```

```python
my_df = my_df.set_index("name")
```

```python
my_df = my_df.loc[(my_df!=0).any(axis=1)]
```

```python
my_df.dropna(axis = 1, inplace = True)
```

```python
clu = sns.clustermap(my_df, cmap = 'bwr', col_cluster = False, yticklabels = True, cbar_pos=(-0.1, 0.2, .03, .4), 
                     standard_scale = 0, figsize= (16, 28), metric='correlation')

for i, ax in enumerate(clu.fig.axes):   ## getting all axes of the fig object|
     ax.set_xticklabels(ax.get_xticklabels(), rotation = 90, fontdict={'fontsize' : 10})
        
for i, ax in enumerate(clu.fig.axes):
    ax.set_yticklabels(ax.get_yticklabels(), fontdict = {'fontsize' : 15})
# plt.savefig('MEBrown_Expression.png')
# plt.show()
# plt.savefig('Tree group.jpg')
# plt.savefig('clustering_carotenoid.png', bbox_inches = "tight")
plt.show()
```
