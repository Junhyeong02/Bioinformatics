```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
```

```python
df = pd.read_csv("MEblack.txt", sep = "\t", index_col = "GeneID")
print(df.head())
```

```python
clu = sns.clustermap(df,  cmap = 'viridis', col_cluster = False, yticklabels = False, standard_scale = 0, cbar_pos = (1.05, 0.2, .03, .4), cbar_kws={"label" : "Nomalized TPM"})

clu.fig.suptitle("MEblack", y = 0.85, fontsize = 25)

print(type(clu.fig))
for i, ax in enumerate(clu.fig.axes):  
     ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)

# plt.savefig('MEBrown_Expression.png')

plt.show()
```

```python
clu = sns.clustermap(df,  cmap = 'viridis', col_cluster = False, yticklabels = False, standard_scale = 0, cbar = False)
plt.title("MEblack")

for i, ax in enumerate(clu.fig.axes):  
     ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)

# plt.savefig('MEBrown_Expression.png')

plt.show()
```
