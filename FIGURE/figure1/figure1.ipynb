```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
```

```python
df = pd.read_csv("MEturquoise.txt", sep = "\t", index_col = "GeneID")
print(df.head())
```

```python
list(df.columns)
```

```python
idx = [
 'ba-root-1',
 'ba-root-2',
 'ba-root-3', 
 'C-ba-Stem-1',
 'C-ba-Stem-2',
 'C-ba-Stem-3', 
 'ba-leaf-1',
 'ba-leaf-2',
 'ba-leaf-3',
 'C-ba-flo-1',
 'C-ba-flo-2',
 'C-ba-Peri-1',
 'C-ba-Peri-2',
 'C-ba-Peri-3',
 'C-ba-ma-peri-1',
 'C-ba-ma-peri-2',
 'C-ba-ma-peri-3',    
 'C-ba-Pla-1',
 'C-ba-Pla-2',
 'C-ba-Pla-3',
 'C-ba-ma-pla-1',
 'C-ba-ma-pla-2',
 'C-ba-ma-pla-3'
]
```

```python
df = df.reindex(idx, axis = 1)
```

```python
df
```

```python
clu = sns.clustermap(df,  cmap = 'seismic', col_cluster = False, yticklabels = False, standard_scale = 0, cbar_pos = (1.05, 0.2, .03, .4), cbar_kws={"label" : "Nomalized TPM"})

clu.fig.suptitle("MEturquoise", y = 0.85, fontsize = 25)

print(type(clu.fig))
for i, ax in enumerate(clu.fig.axes):  
     ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)

# plt.savefig('MEBrown_Expression.png')

plt.show()
```
