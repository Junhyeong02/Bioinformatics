```python
import numpy as np
import pandas as pd
```

```python
path1 = "CABG_MYB.tsv"
```

```python
path2 = "MYB_orthologs.tsv"
```

```python
ortho = pd.read_csv(path2, sep = '\t')
```

```python
ortho.fillna("None", inplace = True)
```

```python
temp = pd.DataFrame()
```

```python
for idx in ortho.index:
    temp.loc[idx, "Orthogroup"] = ortho.loc[idx, "Orthogroup"]
    temp.loc[idx, "CB"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("CB"))))
    temp.loc[idx, "CC"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("CC"))))
    temp.loc[idx, "CM334"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("CaCM334") or i.startswith("CA"))))
    temp.loc[idx, "Chil"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("CaChil") or i.startswith("Capang"))))
    temp.loc[idx, "Zunla-1"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("CaZunla-1") or i.startswith("Capana"))))
    temp.loc[idx, "SF"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("CaSF") or i.startswith("SF"))))
    temp.loc[idx, "ECW"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("CaECW") or i.startswith("ECW"))))
    temp.loc[idx, "Sl"] = ', '.join(list((i for i in list(ortho.loc[idx, "0":]) if i.startswith("SlHeinz") or i.startswith("Solyc"))))
```

```python
temp.replace("", "None", inplace = True)
```

```python
temp.to_csv("orthogroup.tsv", sep = "\t", index = False)
```
