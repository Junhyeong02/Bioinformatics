```python
target_df = pd.read_csv("carotenoid.txt", sep = "\t", index_col = "v2.0")
```

```python
pd.merge(df, target_df, how = "inner", left_index = True, right_index = True)
```

```python
my_df = pd.merge(df, target_df, how = "inner", left_index = True, right_index = True)
```

```python
my_df.set_index("Short name", inplace = True)
```

```python
my_df
```

```python
# my_df.drop(["Short name"], inplace = True, axis = True)
df.columns
```

```python
my_df = my_df.div(my_df.max(axis=1), axis = 0)
```

```python
my_df
```

```python
my_df.dropna(inplace = True)
```

```python
list(my_df.index)
```

```python
index = [
 'PSY1',
 'PDS',    
 'ZDS',
 'LCYB1',
 'BCH1',
 'ZEP2',    
 'VDE',  
 'CCS',
 'NCED4'
]
```

```python
my_df2 = my_df.reindex(index)
```

```python
plt.figure(figsize = (20, 10))
sns.set(font_scale = 1)
ax = sns.heatmap(my_df2, cmap = 'rocket_r')
plt.title('Expression of capsanthin/capsourbin biosynthesis gene', fontsize = 20)
plt.savefig('Expression_of_capsanthin_capsourbin_biosynthesis_gene.pdf')
plt.savefig('Expression_of_capsanthin_capsourbin_biosynthesis_gene.png')
```
