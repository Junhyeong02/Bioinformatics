```python
import sys
import os
import numpy as np
import pandas as pd
```

```python
path = "source/Chi.tissue.TPM.txt"
```

```python
df = pd.read_csv(path, sep = "\t", index_col = 0)
```

```python
df
```

```python
col = ["RT", "ST", "LF", "FL", "PL", "PLB", "PR", "PRB"]
```

```python
for idx in df.index:
    for c in col:
        if c == "PR":
            df.loc[idx, c] = np.mean([df.loc[idx, c + f"-{i}"] for i in range(2, 4)])  

        else:
            df.loc[idx, c] = np.mean([df.loc[idx, c + f"-{i}"] for i in range(1, 4)])
```

```python
df
```

```python
df.to_csv("source/Chi.tissue.avg.tsv", sep = "\t")
```
