```python
path = "tempj.txt"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import rcParams, rc
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
rcParams['text.usetex'] = True
rc('text', usetex = False)
```

```python
df = pd.read_csv(path, sep = "\t")
```

```python
df.drop(9, inplace = True)
```

```python
df
```

```python
fig = plt.figure(figsize= (10, 10))
ax = fig.add_subplot(projection= "3d")

index = len(df.index)

yticks = [
    r"$\it{C. annuum}$ 'CM334'",
    r"$\it{C. annuum}$ 'Chiltepin'",
    r"$\it{C. annuum}$ 'ECW'",
    r"$\it{C. annuum}$ 'SF'",
    r"$\it{C. annuum}$ 'Zunla-1'",
    r"$\it{C. annuum}$ 'Zunla-1' (RefSeq)",
    r"$\it{C. chinense}$ 'PI159236'",
    r"$\it{C. baccatum}$ 'PBC81'",
    r"$\it{S. lycopersicum}$ 'Heinz1706'",
]

width = 0.25
height = 0.6

plt.xticks(np.arange(1, 5), ["1R-MYB", "R2R3-MYB", "3R-MYB", "4R-MYB"], fontsize = 15, 
           horizontalalignment = "center")
plt.yticks(np.arange(index), yticks, verticalalignment = "top", 
           horizontalalignment = "left", fontsize = 12, rotation = -28)

ax.bar3d(1, np.arange(index), 0, width, height, [i for i in df["MYB related"]])
ax.bar3d(2, np.arange(index), 0, width, height, [i for i in df["R2R3 MYB"]])
ax.bar3d(3, np.arange(index), 0, width, height, [i for i in df["3R MYB"]])
ax.bar3d(4, np.arange(index), 0, width, height, [i for i in df["Atypical MYB"]])

plt.xlim(0.5, 5)
ax.set_zlim(0, 250)
ax.set_zticks([0, 50, 100, 150])
ax.set_zticklabels([0, 50, 100, 150], fontdict = {'fontsize' : 15})
ax.set_zlabel("Number of MYB proteins", labelpad = 20, fontdict = {'fontsize': 15})
# ax.set_title("Classification of MYB based on R domain number", fontdict = {'fontsize' : 20})

ax.view_init(22, -40)

for i, val in enumerate(df["MYB related"]):
    ax.text(1.2, i, 250, str(val), fontdict = {'fontsize' : 12, 'color' : 'red'})

for i, val in enumerate(df["R2R3 MYB"]):
    ax.text(2.2, i, 240, str(val), fontdict = {'fontsize' : 12, 'color' : 'red'})

for i, val in enumerate(df["3R MYB"]):
    ax.text(3.2, i, 20, str(val), fontdict = {'fontsize' : 12, 'color' : 'red'})    

for i, val in enumerate(df["Atypical MYB"]):
    ax.text(4.2, i, 20, str(val), fontdict = {'fontsize' : 12, 'color' : 'red'})

plt.savefig("MYB_info.png")
plt.show()
```
