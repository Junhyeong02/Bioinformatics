```python
path = "mapping_data.tsv"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from matplotlib import rcParams, rc
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
rc('text', usetex = False)
```

```python
df = pd.read_csv(path, sep = "\t")
```

```python
sets = list(df["Set"].unique())
```

```python
df_list = []
```

```python
for s in sets:
    df_list.append(df[df["Set"] == s])
```

```python
sorted_df = df_list[4].sort_values("Alignment rate")
```

```python
sorted_df
```

```python
for idx in sorted_df.index:
    sorted_df.loc[idx, "Tissue"] = sorted_df.loc[idx, "Sample"].split("-")[0]
```

```python
list(sorted_df["Tissue"].unique())
```

```python
tissue = [
 'root',    
 'stem',   
 'leaf',
 'flower',
 'placenta1',    
 'placenta2',
 'placenta3',  
 'placentaMG', 
 'placentaB',    
 'placentaB5', 
 'placentaB10',    
 'pericarp1',    
 'pericarp2',    
 'pericarp3',  
 'pericarpMG',    
 'pericarpB',
 'pericarpB5',    
 'pericarpB10'
]
```

```python
my_df = [sorted_df[sorted_df["Tissue"] == t] for t in tissue] 

l = len(my_df)
```

```python
def draw(i, location, df, color_code):
    plt.subplot(*location, sharex = ax1)
    plt.barh(list(df["Sample"]), list(df["Alignment rate"]), 
                   color = color_code, edgecolor = 'white', height=0.7)
    plt.yticks(fontsize = 22, visible = False)
    if i >=3:
        plt.xticks(fontsize = 22, visible = True)
        plt.text(71, -0.25, "70% cutoff", color = "red", fontsize = 25, verticalalignment = "top")
    else:
        plt.xticks(visible = False)
    
    plt.title(df["Tissue"].unique()[0], fontsize = 40)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().margins(y= 0.1)
    plt.axvline(x = 70, color= "red", linewidth = 2)

    for align_rate, sample in zip(list(df["Alignment rate"]), list(df["Sample"])):
        plt.text(align_rate + 1, sample, 
                align_rate, fontsize = 20, verticalalignment = "center")
```
