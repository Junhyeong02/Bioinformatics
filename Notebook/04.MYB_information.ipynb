```python
import sys
import os
import numpy as np
import pandas as pd
```

```python
group_df = pd.read_csv("../data/R2R3_MYB.align.15.group_names.txt", sep = "=")
```

```python
group_df
```

```python
for idx in group_df.index:
    # print(group_df.loc[idx, "Group"])
    group_df.loc[idx, "Gene"] = group_df.loc[idx, "Gene"].replace("_" + group_df.loc[idx, "Group"], "")
    
    if group_df.loc[idx, "Group"].startswith("U"):
        group_df.loc[idx, "Group"] = None
```

```python
group_df
```

```python
group_df["Type"] = "R2R3 MYB"
```

```python
group_df
```

```python
group_df["Organism"] = group_df["Gene"]
```

```python
group_df["Organism"]
```

```python
for idx in group_df.index:
    if group_df.loc[idx, "Organism"].startswith("Am"):
        group_df.loc[idx, "Organism"] = "Antirrhinum majus"

    elif group_df.loc[idx, "Organism"].startswith("At"):
        group_df.loc[idx, "Organism"] = "Arabidopsis thaliana"
        
    elif group_df.loc[idx, "Organism"].startswith("Zm"):
        group_df.loc[idx, "Organism"] = "Zea mays"
        
    elif group_df.loc[idx, "Organism"].startswith("CC"):
        group_df.loc[idx, "Organism"] = "Capsicum chinense"
        
    elif group_df.loc[idx, "Organism"].startswith("CB"):
        group_df.loc[idx, "Organism"] = "Capsicum baccatum"
        
    elif group_df.loc[idx, "Organism"].startswith("Sl"):
        group_df.loc[idx, "Organism"] = "Solanum lycopersicum"
        
    elif group_df.loc[idx, "Organism"].startswith("Ad"):
        group_df.loc[idx, "Organism"] = "Actinidia deliciosa"
        
    else:
        group_df.loc[idx, "Organism"] = "Capsicum annuum"
```

```python
group_df
```

```python
set(group_df["Organism"])
```

```python
group_df = group_df[["Gene", "Organism", 'Type', "Group"]]
```

```python
group_df["temp"] = group_df["Group"]
```

```python
group_df.fillna("U1000", inplace = True)
```

```python
for idx in group_df.index:
    group_df.loc[idx, "temp"] = int(group_df.loc[idx, "temp"][1:])
```

```python
group_df
```

```python
group_df.sort_values(by = "temp", inplace = True)
```

```python
group_df
```

```python
group_df["Group"] = group_df["Group"].replace("U1000", "NA")
```

```python
group_df
```

```python
group_df.drop("temp", axis = 1, inplace = True)
```

```python
group_df
```

```python
group_df.to_csv("../data/R2R3_MYB.group_info.txt", sep = "\t", index = False)
```
